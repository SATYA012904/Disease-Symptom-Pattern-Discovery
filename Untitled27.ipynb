# PCA DIMENSIONALITY REDUCTION

```python
X = df.drop(columns=['diseases'])
y = df['diseases']
```

```python
X.shape
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    cumulative_variance,
    marker='o'
)
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.grid()
plt.show()
```

```python
n_components = np.argmax(cumulative_variance >= 0.80) + 1
```

```python
n_components
```

```python
pca_final = PCA(n_components=n_components)
pca_final_data = pca_final.fit_transform(scaled_data)
print("\nFinal PCA Shape:")
print(pca_final_data.shape)
```

```python
pca_df = pd.DataFrame(pca_final_data)
X_pca = pca_final_data
```

```python
inertia = []
K = range(2, 15)

for k in K:

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.utils import shuffle

# ------------------------------------------
# TAKE RANDOM SAMPLE
# ------------------------------------------

X_sample = shuffle(
    X_pca,
    random_state=42
)[:20000]

print("Sample Shape:")
print(X_sample.shape)

# ------------------------------------------
# CALCULATE SCORES
# ------------------------------------------

silhouette_scores = []

K = [5,7,10,12,14]

for k in K:

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X_sample)

    score = silhouette_score(
        X_sample,
        labels
    )

    silhouette_scores.append(score)

# ------------------------------------------
# PRINT SCORES
# ------------------------------------------

print("\nSilhouette Scores:\n")

for k, score in zip(K, silhouette_scores):
    print(f"K = {k} --> Score = {score:.4f}")
```

#  CLUSTERING

```python
kmeans = KMeans(
    n_clusters=12,
    random_state=42,
    n_init=10
)
```

```python
clusters = kmeans.fit_predict(X_pca)
print("K-Means Clustering Completed")
```

```python
df['Cluster'] = clusters
```

```python
print("\nCluster Distribution:\n")
print(df['Cluster'].value_counts())
```

```python
feature_columns = df.drop(
    columns=['diseases', 'Cluster']
).columns
```

```python
for cluster_num in sorted(df['Cluster'].unique()):

    print("\n")
    print("="*50)
    print(f"CLUSTER {cluster_num}")
    print("="*50)

    cluster_num = df[
        df['Cluster'] == cluster_num
    ]

    # Mean symptom frequency
    symptom_means = cluster_num[
        feature_columns
    ].mean()

    # Top 10 symptoms
    top_symptoms = symptom_means.sort_values(
        ascending=False
    ).head(10)

    print("\nTop Symptoms:\n")

    print(top_symptoms)
```

```python
pca_2d = PCA(n_components=2)

X_2d = pca_2d.fit_transform(X_pca)

print("2D PCA Shape:")
print(X_2d.shape)
```

```python
scatter = plt.scatter(
    X_2d[:,0],
    X_2d[:,1],
    c=clusters,
    cmap='tab20',
    s=10,
    alpha=0.7
)

# Labels
plt.title("Disease Symptom Pattern Clusters", fontsize=16)

plt.xlabel("PCA Component 1")

plt.ylabel("PCA Component 2")

# ------------------------------------------
# CREATE LEGEND
# ------------------------------------------

legend = plt.legend(
    *scatter.legend_elements(),
    title="Clusters",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.gca().add_artist(legend)

# Show plot
plt.show()
```

```python
cluster_names = {
    0: "Pediatric/Viral",
    1: "Urinary-Prostate",
    2: "Psychiatric",
    3: "Cardio-Respiratory",
    4: "Musculoskeletal",
    5: "ENT Disorders",
    6: "Gastrointestinal",
    7: "Urinary Infection",
    8: "Eye Disorders",
    9: "Gynecological",
    10: "Respiratory Infection",
    11: "Skin Disorders"
}
```

```python
cmap = plt.get_cmap('tab20')

# Create figure
plt.figure(figsize=(16,12))

# Scatter plot
scatter = plt.scatter(
    X_2d[:,0],
    X_2d[:,1],
    c=clusters,
    cmap=cmap,
    s=10,
    alpha=0.7
)

# ------------------------------------------
# CREATE CUSTOM LEGEND WITH COLORS
# ------------------------------------------

handles = []

for cluster_id in np.unique(clusters):

    color = cmap(cluster_id)

    handles.append(
        plt.Line2D(
            [],
            [],
            marker='o',
            linestyle='',
            markersize=10,
            color=color,
            label=f"{cluster_id} : {cluster_names[cluster_id]}"
        )
    )

# Add legend
plt.legend(
    handles=handles,
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    title="Disease Clusters"
)

# Titles
plt.title(
    "Disease Symptom Pattern Clusters",
    fontsize=18
)

plt.xlabel("PCA Component 1")

plt.ylabel("PCA Component 2")

plt.show()
```
