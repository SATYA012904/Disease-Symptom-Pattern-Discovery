# PCA DIMENSIONALITY REDUCTION

```python
X = df.drop(columns=['diseases'])
y = df['diseases']
```

```python
X.shape
```

```python
print(X.isnull().sum().sum())
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python

pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    cumulative_variance,
    marker='o'
)
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.grid()
plt.show()
```

```python
n_components = np.argmax(cumulative_variance >= 0.80) + 1
```

```python
n_components
```

```python
pca_final = PCA(n_components=n_components)
pca_final_data = pca_final.fit_transform(scaled_data)
print("\nFinal PCA Shape:")
print(pca_final_data.shape)

import joblib
joblib.dump(pca_final, "pca_model.pkl")
```

```python
pca_df = pd.DataFrame(pca_final_data)
X_pca = pca_final_data
```

```python
inertia = []
K = range(2, 15)

for k in K:

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.grid()
plt.show()
```

#  CLUSTERING

```python
kmeans = KMeans(
    n_clusters=13,
    random_state=42,
    n_init=10
)
```

```python
clusters = kmeans.fit_predict(X_pca)
print("K-Means Clustering Completed")

import joblib
joblib.dump(kmeans, "kmeans_model.pkl")
```

```python
df['Cluster'] = clusters
```

```python
print("\nCluster Distribution:\n")
print(df['Cluster'].value_counts())
```

```python
feature_columns = df.drop(
    columns=['diseases', 'Cluster']
).columns
```

```python
for cluster_num in sorted(df['Cluster'].unique()):

    print("\n")
    print("="*50)
    print(f"CLUSTER {cluster_num}")
    print("="*50)

    cluster_num = df[
        df['Cluster'] == cluster_num
    ]

    # Mean symptom frequency
    symptom_means = cluster_num[
        feature_columns
    ].mean()

    # Top 10 symptoms
    top_symptoms = symptom_means.sort_values(
        ascending=False
    ).head(10)

    print("\nTop Symptoms:\n")

    print(top_symptoms)
```

```python
pca_2d = PCA(n_components=2)

X_2d = pca_2d.fit_transform(X_pca)

print("2D PCA Shape:")
print(X_2d.shape)
```

```python
import pandas as pd

patient_df = pd.read_csv(
    "final_patient_symptoms.csv"
)

patient_df.head()
```

```python
import joblib

scaler = joblib.load("scaler.pkl")

pca = joblib.load("pca_model.pkl")

kmeans = joblib.load("kmeans_model.pkl")

symptom_columns = joblib.load(
    "symptom_columns.pkl"
)
```

```python
input_df = pd.DataFrame(
    0,
    index=range(len(patient_df)),
    columns=symptom_columns
)
```

```python
symptom_cols = [
    'Symptom_1',
    'Symptom_2',
    'Symptom_3',
    'Symptom_5',
    'Symptom_6',
    'Symptom_7',
    'Symptom_8',
    'Symptom_9',
    'Symptom_10',
    'Symptom_11',
    'Symptom_12'
]

for i, row in patient_df.iterrows():

    for col in symptom_cols:

        symptom = str(row[col]).strip().lower()

        # Match symptom column
        if symptom in input_df.columns:

            input_df.loc[i, symptom] = 1
```

```python
scaled_input = scaler.transform(
    input_df
)
```

```python
pca_input = pca.transform(
    scaled_input
)
```

```python
predicted_clusters = kmeans.predict(
    pca_input
)
```

```python
cluster_names = {

    0: "Gynecological & Pregnancy Disorders",

    1: "Spine & Nerve Pain Disorders",

    2: "Eye Infection & Conjunctival Disorders",

    3: "Gastrointestinal & Metabolic Disorders",

    4: "Dental & Oral Disorders",

    5: "Vision & Retinal Disorders",

    6: "Musculoskeletal & Joint Disorders",

    7: "ENT & Ear Disorders",

    8: "Common Viral & Gastrointestinal Infections",

    9: "Respiratory & Pulmonary Disorders",

    10: "Urinary & Prostate Disorders",

    11: "Psychiatric & Behavioral Disorders",

    12: "Skin & Dermatological Disorders"
}
```

```python
patient_df['Predicted_Cluster'] = (
    predicted_clusters
)

patient_df['Disease_Category'] = (
    patient_df['Predicted_Cluster']
    .map(cluster_names)
)

print(
    patient_df[
        ['patient', 'Disease_Category']
    ]
)
```

```python
matched_count = []

for i, row in patient_df.iterrows():

    count = 0

    for col in symptom_cols:

        symptom = str(row[col]).strip().lower()

        if symptom in input_df.columns:

            count += 1

    matched_count.append(count)

patient_df['Matched_Symptoms'] = matched_count

print(
    patient_df[
        ['patient', 'Matched_Symptoms']
    ]
)
```
